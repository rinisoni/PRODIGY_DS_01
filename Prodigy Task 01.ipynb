```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('world_population.csv')#dataset from kaggle
plt.figure(figsize=(12,5))

#histogram for continuous value (2022 population)
plt.hist(df['2022 Population'], bins=10,color="skyblue",edgecolor="black")
plt.xlabel('2022 Population')
plt.ylabel('Frequency')
plt.title('Histogram of Population Distributution for the year 2022')
```

```python
#bar chart for average population by continent (categorical value)
population_continent=df.groupby('Continent')['2022 Population'].mean()#calculate mean population by continent
population_continent.plot(kind='bar',color=['orange','green'])
plt.xlabel('Continent')
plt.ylabel('Average Population')
plt.title('Average Population by Continent')
plt.show()
```
