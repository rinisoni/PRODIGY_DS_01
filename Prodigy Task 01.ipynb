```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data_file_path = r'C:\Users\RINI\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_3401680\Population dataset.csv'
population_data = pd.read_csv(data_file_path, skiprows=4)

# Select data for the year 2023
year = '2023'
population_2023 = population_data[['Country Name', year]].dropna()

# Sort the data by population size (descending order)
population_2023_sorted = population_2023.sort_values(by=year, ascending=False).head(10)  # Top 10 countries by population

# Plotting the bar graph
plt.figure(figsize=(14, 8))
plt.barh(population_2023_sorted['Country Name'], population_2023_sorted[year], color='skyblue')
plt.xlabel('Population in 2023')
plt.ylabel('Country Name')
plt.title('Top 10 Countries by Population in 2023')
#plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```
